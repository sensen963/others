```python
# ノードの大きさを調整する関数の例
def log_700(x):
    return np.log(x)*700
def mul_700(x):
    return x*700
```

```python
# グラフを描画．
# 引数は隣接リスト，表示するノード数,重み関数,ノード間の反発力,書き出すファイル名(今回は書き出さないので'NONE')
OutputNGraph(network_list1, len(network_list1), mul_700, 0.7, 'NONE')
```
