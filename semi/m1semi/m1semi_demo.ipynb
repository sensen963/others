```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

#パラメータ(lambda>k^-1)

mu = 1
lam = 5
k = 3

#初期値
v0 = [0.9999, 0.0001, 0.0]

t = np.arange(0, 20, 1)

#関数
def func(v, t, mu, lam, k):
    return [-lam*k*v[1]*v[0], -mu*v[1]+lam*k*v[1]*v[0], mu*v[1]]

v = odeint(func, v0, t, args=(mu, lam, k))

plt.plot(t, v[:,0], label = "s(t)")
plt.plot(t, v[:,1], label = "rho(t)")
plt.plot(t, v[:,2], label = "r(t)")

plt.legend()
plt.show()
```
